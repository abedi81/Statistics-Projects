```python
import numpy as np
import statsmodels.formula.api as smf
import pandas as pd 
import statsmodels.api as sm
```

```python
# Simulating or generating obeseravtions or X 
np.random.seed(1)
x = np.random.uniform(0, 10, size=500)
mean_x = x.mean()
std_x = x.std(ddof=1)  

print(f"Sample mean of x: {mean_x:.2f}")
print(f"Sample standard deviation of x: {std_x:.2f}")
```

```python
# Simulating or generating error or u
u = np.random.normal(0, 1, size=500) * 6  
mean_u = u.mean()
std_u = u.std(ddof=1)

print(f"Sample mean of u: {mean_u:.2f}")
print(f"Sample standard deviation of u: {std_u:.2f}")
```

```python
# one of the important condition of OLS (Ordinary Least Square) is the sum of error should be zero .As we can see 
# because the values are random samples from the distribution with only 500 draws, the sample mean will approximate the
# true mean (0), but it will almost never be exactly 0 unless we make the sample large enough. It is where Law of Large 
# Number comes up and saying as the number of observations gets larger (by the Law of Large Numbers), the sample mean will 
# get closer and closer to 0.
```

```python
# Now we want to simulate     y_i=1+2xi+Ui=b0+b1.xi+ui
# the equation in the left side is the true model and we want to estimate it by the equation in the right side through 
# data generation process that we have above. 
```

```python
y = 1 + 2 * x + u

X = sm.add_constant(x)  
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
#  As we can see The estimated intercept (β̂₀ ≈ 0.8) and slope (β̂₁ ≈ 2.01) are very close to the true population values
# (1 and 2). They are not exactly equal because the sample includes random error (uᵢ).But if we repeated this experiment
# many times, the average of the estimates across simulations would be: 𝐸[𝛽^0]=1 and 𝐸[𝛽^1]=2. 
# Its excatly the definition of OLS being unbiased.The estimates we get from one sample won’t exactly match the true 
# parameters (because of randomness in uᵢ), but on average I mean across many samples they’ll be equal to the population values.
```

```python
np.random.seed(1)
x = np.random.uniform(0, 10, 100)
u = np.random.normal(0, 2, 100)
y = 1 + 2 * x + u

X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

u_hat = model.resid
x_with_const = X  

sum_u = np.sum(u_hat)
sum_xu = np.sum(x * u_hat)

print("Sum of residuals:", round(sum_u, 10))
print("Sum of x_i * residuals:", round(sum_xu, 10))
```

```python
# Through above coding, it shows the imporatnt propertise of resiaul whih is 𝑢^𝑖=𝑦i−𝑦^𝑖

# The sum of residuals equals zero: ∑𝑢^𝑖=0
# The sum of residuals weighted by x equals zero:∑𝑥𝑖*𝑢^𝑖=0


# These come directly from the first-order conditions of minimizing the sum of squared residuals.
# The first condition ensures the regression line passes through the mean of the data. In other word, there are sone 
# unobservable factors , but these factors cancel each other out when they sum up. 
# The second condition ensures that residuals are uncorrelated with x which means x explains all systematic variation 
# in y that OLS can find.In other word, it is saying that there is no other factors in residual ( which is unobeserved 
# factors) or we accounted for them or control them. Put it better way. There is no Omitted Variable to cause bias. 
```

```python
# Now we compare the error with residual

np.random.seed(1)
n = 100
x = np.random.uniform(0, 10, n)
u = np.random.normal(0, 2, n)
y = 1 + 2*x + u 


X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

u_hat = model.resid   # residuals
u_true = u            # true errors

# for OLS 
sum_resid = np.sum(u_hat)
sum_x_resid = np.sum(x * u_hat)

#  for true errors
sum_errors = np.sum(u_true)
sum_x_errors = np.sum(x * u_true)

print("For residuals:")
print("∑ ûi =", round(sum_resid, 5))
print("∑ xiûi =", round(sum_x_resid, 5))

print("\nFor true errors:")
print("∑ ui =", round(sum_errors, 5))
print("∑ xiui =", round(sum_x_errors, 5))
```

```python
# In the true population model(𝑦𝑖=𝛽0+𝛽1.𝑥𝑖+𝑢𝑖) we have the true errors (the unobserved random noise) and we calculate it
# like ui=yi−(β0+β1*xi)

# When we estimate the model from sample data, we get residuals(𝑢^𝑖=𝑦𝑖−𝑦^𝑖) not error. The residuals are constructed to 
# satisfy the two basic conditions we have above and we force it to sum of it and sum of its weight to be zero and make 
# the line fit the data as best as possible by choosing 𝛽^0 and 𝛽^1 to make the total squared residuals as small as 
# possible.

# But true errors (uᵢ) are random “mistakes” in the real world.Nature doesn’t care about making them add up to zero.
# They might be mostly positive in one sample and mostly negative in another.On average (over many samples), their mean is
# 0,but in one random sample, they won’t cancel exactly. In other word we need a very big sample to make it close to zero.
```

```python
# now we repaet the same process above to see what we can get 

np.random.seed(2)    
x = np.random.uniform(0, 10, 500)

u = np.random.normal(0, 6, 500)

y = 1 + 2*x + u

X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# as we can in above table we got different values for the parameters this time. Even though the true model is the same
# the sample changes slightly due to random variation.OLS estimates depend on the particular sample so β̂₀ and β̂₁ will
# fluctuate around the true values.But again across many repetitions, the average estimates will approach the true 
# population values (1 and 2).
```
